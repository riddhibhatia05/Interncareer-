```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
x = iris.data
y = iris.target

print(x)
```

```python
print(y)
```

```python
np.mean(x)
```

```python
from scipy.stats import norm
```

```python
def histogram(name,data):
    mean = np.mean(data)
    sd = np.std(data)
    x_axis = np.linspace(min(data), max(data),100)
    plt.plot(x_axis,norm.pdf(x_axis, mean, sd), color="blue", label="Normal Distribution")
    plt.xlabel(f'{name}')
    plt.ylabel('density')
    plt.title(f'Normal Distribution - {name}')
    plt.legend()
    plt.show()
    
    plt.hist(data, bins = 20, density=True, alpha=0.6, label = "Data Histogram")
    plt.xlabel(f'{name}')
    plt.ylabel('density')
    plt.title(f'Histogram - {name}')
    plt.legend()
    plt.show()
    
sepal_length = iris.data[:, 0]
sepal_width = iris.data[:, 1]
petal_length = iris.data[:, 2]
petal_width = iris.data[:, 3]

histogram("Sepal Length", sepal_length)
histogram("Sepal Width", sepal_width)
histogram("Petal Length", petal_length)
histogram("Petal Width", petal_width)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

np.mean(x_train_scaled)
```

```python
lr_model = LogisticRegression()
lr_model.fit(x_train_scaled, y_train)
lr_predictions = lr_model.predict(x_test_scaled)
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train_scaled, y_train)
dt_prediction = dt_model.predict(x_test_scaled)
```

```python
svm_model = SVC()
svm_model.fit(x_train_scaled, y_train)
svm_prediction = svm_model.predict(x_test_scaled)
```

```python
def evaluate_model(model_name, prediction):
    accuracy = accuracy_score(y_test, prediction)
    class_report = classification_report(y_test, prediction, target_names = iris.target_names)
    
    print(f"{model_name} Model")
    print("Accuracy: ", accuracy)
    print("Classification Report:\n ", class_report)
    
evaluate_model("Logistic Regression", lr_predictions)
evaluate_model("Support Vector Machine", svm_prediction)
evaluate_model("Decision Tree", dt_prediction)
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
def plot_confusion_matrix(model_name, prediction):
    cm = confusion_matrix(y_test, prediction)
    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt="d",cmap="Blues",xticklabels = iris.target_names, yticklabels = iris.target_names)
    plt.title(f"{model_name} Confusion Matrix")
    plt.xlabel("Predicated Lables")
    plt.ylabel("True Lables")
    plt.show()
    
plot_confusion_matrix("Logistic Regression", lr_predictions)
plot_confusion_matrix("Support Vector Machine", svm_prediction)
plot_confusion_matrix("Decision Tree", dt_prediction)
```

```python
def plot_accuracy_comparison(models, accuracies):
    plt.figure(figsize=(10,6))
    plt.bar(models, accuracies, color='skyblue')
    plt.title("Model Accuracy Comparison")
    plt.xlabel("Model")
    plt.ylabel("Accuracy")
    plt.ylim(0.8, 1.0)
    plt.show()
    
model_names = ["Logistic Regression", "Support Vector Machine", "Decision Tree"]
model_accuracies = [accuracy_score(y_test, lr_predictions),
                  accuracy_score(y_test, svm_prediction),
                  accuracy_score(y_test, dt_prediction)]

plot_accuracy_comparison(model_names, model_accuracies)
```
